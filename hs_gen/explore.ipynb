```python
bitpat_mnemonic_pairs = list(zip(isa.isa_packed_df.Mnemonic, isa.isa_packed_df.PAGE))
opcodes = []

for index, (mnemonic, page_no) in enumerate(zip(isa.isa_packed_df.Mnemonic, isa.isa_packed_df.PAGE)):
    if index == 0:
        opcodes += [f"  = {mnemonic:<23} -- page {page_no}"]
    else:
        opcodes += [f"  | {mnemonic:<23} -- page {page_no}"]

content = \
f'''{{-# LANGUAGE DataKinds #-}}
{{-# LANGUAGE NumericUnderscores #-}}

module Opcodes(Opcode(..)) where
import Clash.Prelude

data Opcode
{'\n'.join(opcodes)}
  deriving (Generic, Show, Eq, NFDataX)
'''
```
